```python
import scipy as sp
import scipy.stats
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
N = 30
mu = [0.1, 0.1, 0.1, 0.1, 0.3, 0.3]
rv = sp.stats.multinomial(N, mu)

np.random.seed(0)
X = rv.rvs(100)
X[:10]
```

```python
df = pd.DataFrame(X).stack().reset_index()
df.columns = ["", "", ""]

sns.violinplot(x="", y="", data=df, inner="quartile")
sns.swarmplot(x="", y="", data=df, color=".3")
plt.title("  ")
plt.show()
```

```python
plt.boxplot(X)
plt.title("  ")
plt.xlabel("")
plt.ylabel("")
plt.show()
```
